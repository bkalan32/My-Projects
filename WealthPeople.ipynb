```python
# Importing Libraries
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

### Prepare data

```python
# Read in data "forbes_richman.csv" file into pandas Dataframe
wealth = pd.read_csv('forbes_richman.csv', encoding='ISO-8859-1', index_col='Rank')

# Show the Dataframe
wealth
```

# Remove the columns that contain missing values

```python
# How many null values in each column
wealth.isnull().sum()
```

```python
# Drop Null values
wealth.dropna(inplace=True)

wealth
```

```python
wealth.isnull().sum()
```

### Find everyone into tech that's 50 or younger.

```python
# Find how many Wealthy people are in the Technology industry and plot
technology = wealth[wealth['Industry']=='Technology']

# Select the age younger or equal to 50
tech_age = technology[technology['Age']<=50]

# Display the data
display(tech_age.head())
display(tech_age.tail())
```

```python
tech_age.shape
```

### Find everyone thats into Fashion & Retail that's 50 or younger.

```python
# Find how many Fashion people are in the Technology industry and plot
Fashion = wealth[wealth['Industry']=='Fashion & Retail']

# Select the age younger or equal to 50
fashion_age = Fashion[Fashion['Age']<=50]

# Display the data
display(fashion_age.head())
display(fashion_age.tail())
```

```python
fashion_age.shape
```

### Combine both dataframes


```python
wealth_comb = pd.concat([fashion_age,tech_age])
wealth_comb
```

```python
# Create a bar chart for Fashion industry
fashion_count = fashion_age.shape[0]
plt.bar(["Fashion"], [fashion_count], color='blue')

# Create a bar chart for Technology industry
tech_count = tech_age.shape[0]
plt.bar(["Technology"], [tech_count], color='green')

# Set the plot title and axis labels
plt.title("Number of people in Fashion and Technology industries under 50 years old")
plt.xlabel("Industry")
plt.ylabel("Number of People")

# Display the plot
plt.show()
```

```python
# Display using a pie chart

# Create data
counts = [fashion_age.shape[0], tech_age.shape[0]]
labels = ["Fashion", "Technology"]
colors = ["blue", "green"]

# Create pie chart
plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%')

# Set the plot title
plt.title("Number of people in Fashion and Technology industries under 50 years old")

# Display the plot
plt.show()
```
